```python
# Define weights
w0 = -1
w1 = 0

#Define input vectors
x0 = [0, 1,0,1]
x1 = [0, 0,1,1]
```

```python
for val_x0, val_x1 in zip(x0, x1):
    y0 = (val_x0 * w0) + (val_x1 * w1)

  #(if sum is 0 or more, output is 1)
    if y0 >= 0:
        output = 1
    else:
        output = 0
#Print the result for each iteration
    print(output)
```
