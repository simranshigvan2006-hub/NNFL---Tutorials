```python
#Define weights
w0 = 1
w1 = 1

#Define input vectors (Truth Table for OR gate)
x0 = [0, 0, 1, 1]
x1 = [0, 1, 0, 1]
```

```python
for val_x0, val_x1 in zip(x0, x1):
    y0 = (val_x0 * w0) + (val_x1 * w1)

#(if sum is 1 or more, output is 1)
    if y0 >= 1:
        output = 1
    else:
        output = 0
#Print the result for each iteration
    print(output)
```
